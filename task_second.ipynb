1) Пусть $f(x) = 2\sin x - \cos{2x}$. Напишите функцию `beauty_derivative_plot`, которая берет точку $x$ и строит график функции $f(x)$ и ее производных $f'(x)$ в диапазоне $x\in [-2\pi, 2\pi]$. Функция также должна построить касательную к функции в точке $x$.


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2 * np.sin(x) - np.cos(2 * x)

def df(x):
    return 2 * np.cos(x) + 2 * np.sin(2 * x)

def beauty_derivative_plot(x_point):
    x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
    y = f(x)
    dy = df(x)

    y_point = f(x_point)
    dy_point = df(x_point)

    x_tangent = np.linspace(x_point - 1, x_point + 1, 100)
    y_tangent = dy_point * (x_tangent - x_point) + y_point

    plt.figure(figsize = (15, 5))
    plt.plot(x, y, label = r'$f(x) = 2\sin(x) - \cos(2x)$', color = 'blue')
    plt.plot(x, dy, label = "$f'(x) = 2\cos(x) + 2\sin(2x)$", linestyle = 'dashed', color = 'red')
    plt.plot(x_tangent, y_tangent, label = f'Tangent at $x = {x_point:.2f}$', color = 'green')

    plt.scatter([x_point], [y_point], color = 'black', zorder = 3)

    plt.vlines(x_point, ymin = 0, ymax = y_point, color = 'purple', linestyle = 'dashed')
    plt.hlines(y_point, xmin = 0, xmax = x_point, color = 'purple', linestyle = 'dashed')

    plt.annotate(f"f'({x_point:.2f}) = {dy_point:.2f}", xy=(x_point, y_point), xytext=(x_point - 1, y_point - 2),
                 arrowprops=dict(facecolor='black', arrowstyle='->', connectionstyle="arc3,rad=0.2222"),
                 fontsize = 10)

    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.title('Function, Derivative, and Tangent Line')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Примеры использования:
beauty_derivative_plot(-3)
beauty_derivative_plot(-2)
beauty_derivative_plot(1.714)
```
